```python
def convert_sqft_to_num(x):
  # If it's already a number (int or float), return it as a float
  if isinstance(x, (int, float)):
    return float(x)

  # Otherwise, assume it's a string and try to process it
  try:
    tokens = str(x).split('-')
    if len(tokens) == 2:
      return (float(tokens[0]) + float(tokens[1])) / 2
  except:
    # If splitting or converting tokens fails, return None
    return None

  # If it's not a range, try to convert the whole string to a float
  try:
    return float(x)
  except:
    return None

df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
df3.head(20)
```

```python
def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<(m+st))]
    df_out = pd.concat([df_out, reduced_df], ignore_index=True)
  return df_out
```

```python
def REMOVE_BHK_outlier(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.std(bhk_df.price_per_sqft), # Corrected column name to price_per_sqft
          'count': bhk_df.shape[0]
      }
    # Now, iterate through the bhk groups again for this location to find outliers
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk - 1)
      if stats and stats['count'] > 5: # Check if stats for previous bhk exist and are reliable
        # Identify outliers where current bhk's price_per_sqft is less than mean of bhk-1
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
  # Return outside all loops to process the entire DataFrame
  return df.drop(exclude_indices, axis="index")
```

```python
df7 = REMOVE_BHK_outlier(df6)
df7.shape
```
